```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cardio_train.csv",sep=";")
df
```

```python
df = df.drop('id', axis = 1)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.cardio.value_counts()
```

```python
df['age_years'] = df['age'] // 365 
df = df.drop('age', axis = 1)  
```

```python
df
```

```python
df['bmi'] = df['weight'] / ((df['height'] / 100) ** 2)
```

```python
numerical_cols = ['age_years', 'height', 'weight', 'ap_hi', 'ap_lo','bmi']
df[numerical_cols].hist(bins=20, figsize=(12, 8))
plt.show()
```

```python
# Impossible values
# ap_hi is systolic bp: max bp 
# ap_lo is diastolic bp: min bp

df = df[(df['ap_hi'] >= 70) & (df['ap_hi'] <= 240)]
df = df[(df['ap_lo'] >= 40) & (df['ap_lo'] <= 150)]
```

```python
# diastolic > systolic not possible 

df = df[df['ap_hi'] > df['ap_lo']]
```

```python
# BP = 0 not possible 

df = df[(df['ap_hi'] != 0) & (df['ap_lo'] != 0)]
```

```python
df = df[df['ap_hi'].between(90, 180)]      
df = df[df['ap_lo'].between(60, 120)]      
df = df[df['height'].between(140, 200)]
df = df[df['weight'].between(45, 150)]
# df = df[df['bmi'] < 50]
```

```python
numerical_cols = ['age_years', 'height', 'weight', 'ap_hi', 'ap_lo','bmi']
df[numerical_cols].hist(bins=20, figsize=(12, 8))
plt.show()
```

```python
for col in numerical_cols:
    sns.boxplot(x='cardio', y=col, data=df)
    plt.title(f'{col} by Cardio')
    plt.show()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('cardio', axis=1)
y = df['cardio']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
#Boosting

from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

params = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1]
}

grid = GridSearchCV(
    AdaBoostClassifier(random_state=42),
    params,
    cv=5,
    scoring='f1',
    n_jobs=-1
)

grid.fit(X_train, y_train)

print(grid.best_params_)
print(classification_report(y_test, grid.predict(X_test)))

#Accuracy => ~73%
```

```python
# Linear SVM

from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', LinearSVC(max_iter=5000))
])

params = {
    'svm__C': [0.01, 0.1, 1, 10]
}

grid = GridSearchCV(
    pipe,
    params,
    cv=5,
    scoring='f1'
)

grid.fit(X_train, y_train)

print(grid.best_params_)
print(classification_report(y_test, grid.predict(X_test)))

#Accuracy=> ~73%
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

params = {
    'n_estimators': [50, 100, 200],
    'max_samples': [0.7, 1.0],
    'max_features': [0.7, 1.0]
}

bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=42),
    random_state=42
)

grid = GridSearchCV(
    bagging,
    params,
    cv=5,
    scoring='f1',
    n_jobs=-1,     
    verbose=2 
)

grid.fit(X_train, y_train)

print("Best Params:", grid.best_params_)
y_pred = grid.predict(X_test)
print(classification_report(y_test, y_pred))

#Accuracy => ~71%
```

```python
#Decision Tree

from sklearn.model_selection import GridSearchCV

params = {
    "max_depth": [None, 5, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 5],
    "criterion": ["gini", "entropy"]
}

grid = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    params,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

grid.fit(X_train, y_train)

print("Best Params:", grid.best_params_)
print(classification_report(y_test, grid.predict(X_test)))

#Accuracy=> ~72%
```

```python
#KNN

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5) 

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("KNN Test Accuracy:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

#Accuracy=> ~68.27%
```

```python
#K-Fold

from sklearn.model_selection import KFold, cross_val_score


kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(knn, X, y, cv=kf, scoring='accuracy')
print("Accuracy for each fold:", scores)

print("Mean Accuracy:", np.mean(scores))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

#Accuarcy => ~68.51
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

rf = RandomForestClassifier(
    n_estimators=500,          
    max_depth=10,              
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)   

y_pred_rf = rf.predict(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", round(acc_rf, 4))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
